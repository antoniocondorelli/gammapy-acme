## 2. Apply Quality Cuts

```python
print("Applying cuts...")
q_mask_nu = get_cuts(df_nu.bdt0, df_nu.bdt1, df_nu.dir_z)
df_nu_q = df_nu[q_mask_nu].copy()
q_mask_nubar = get_cuts(df_nubar.bdt0, df_nubar.bdt1, df_nubar.dir_z)
df_nubar_q = df_nubar[q_mask_nubar].copy()
q_mask_mu = get_cuts(df_mu.bdt0, df_mu.bdt1, df_mu.dir_z)
df_mu_q = df_mu[q_mask_mu]

print("nu: {:d} events survive cuts ({:.4g}%)".format(len(df_nu_q), len(df_nu_q)/len(df_nu)*100))
print("nubar: {:d} events survive cuts ({:.4g}%)".format(len(df_nubar_q), len(df_nubar_q)/len(df_nubar)*100))
print("{:d} mu events survive cuts ({:.4g}%)".format(len(df_mu_q), len(df_mu_q)/len(df_mu)*100))
```

## 8. Compute Atmospheric Backgrounds

```python
print(" ")
print("---- Start Background calculation---")
#---------Binning -------
cos_bins_fine = np.linspace(1, -1, 13)
t_bins_fine = np.arccos(cos_bins_fine)
e_bins_fine = np.logspace(2, 8, 49)

cos_bins_coarse = np.linspace(1, -1, 7)
t_bins_coarse = np.arccos(cos_bins_coarse)
e_bins_coarse = np.logspace(2, 8, 25)

migra_bins = np.logspace(-5, 2, 57)
rad_bins = np.concatenate((np.linspace(0,1,20, endpoint=False), np.linspace(1,5,40, endpoint=False), np.linspace(5,30,51), [180.0]))
# Bin centers
e_binc_fine = np.sqrt(e_bins_fine[:-1] * e_bins_fine[1:])
e_binc_coarse = np.sqrt(e_bins_coarse[:-1] * e_bins_coarse[1:])

t_binc_fine = np.arccos(0.5*(cos_bins_fine[:-1] + cos_bins_fine[1:]))
t_binc_coarse = np.arccos(0.5*(cos_bins_coarse[:-1] + cos_bins_coarse[1:]))

migra_binc = np.sqrt(migra_bins[:-1] * migra_bins[1:])
rad_binc = 0.5*(rad_bins[1:] + rad_bins[:-1])
sizes_migra_bins = np.diff(migra_bins)
e_bins_fine_ext = np.logspace(0, 8, 65)
e_binc_fine_ext = np.sqrt(e_bins_fine_ext[:-1] * e_bins_fine_ext[1:])

assert (e_bins_fine_ext[16:] == e_bins_fine).all()

#----------Load knee corrections---------
print("Loading knee corrections...")
honda_knee_e, honda_knee_f = np.loadtxt('knee_corrections/honda_knee_correction.dat', unpack=True)
enberg_knee_e, enberg_knee_f = np.loadtxt('knee_corrections/enberg_knee_correction.dat', unpack=True)

# log10 is used as the knee feature is typically better behaved in log-space
honda_knee_correction = lambda e:np.interp(np.log10(e), np.log10(honda_knee_e), honda_knee_f)
enberg_knee_correction = lambda e:np.interp(np.log10(e), np.log10(enberg_knee_e), enberg_knee_f)

#Dictionaries to store atmospheric fluxes (conventional and prompt)
atm_conv_flux = dict()
atm_prompt_flux = dict()

#Initialize 2D arrays (Energy x Zenith) for fluxes
atm_conv_flux['nu'] = np.zeros((len(e_binc_fine), len(t_binc_fine)))
atm_conv_flux['nubar'] = np.zeros((len(e_binc_fine), len(t_binc_fine)))
atm_prompt_flux['nu'] = np.zeros((len(e_binc_fine), len(t_binc_fine)))
atm_prompt_flux['nubar'] = np.zeros((len(e_binc_fine), len(t_binc_fine)))

print("Calculating atmospheric neutrino fluxes and applying knee corrections...")
for l,pt in zip(['nu', 'nubar'], [14, -14]):
    for i in range(len(t_binc_fine)):
        # retrieve atmospheric fluxes for each energy and zenith bin center
        # conventional flux calculation
        conv_flux = np.array([km3_flux.atmospheric_conventional(mc_type=pt, zenith=t_binc_fine[i], energy=e) \
                              for e in e_binc_fine])

        # prompt flux calculation
        prompt_flux = np.array([km3_flux.atmospheric_prompt(mc_type=pt, zenith=t_binc_fine[i], energy=e) \
                               for e in e_binc_fine])

        # apply knee corrections: Honda for conventional, Enberg for prompt
        conv_flux *= honda_knee_correction(e_binc_fine)
        prompt_flux *= enberg_knee_correction(e_binc_fine)

        # Store the calculated and corrected fluxes
        atm_conv_flux[l][:,i] = conv_flux
        atm_prompt_flux[l][:,i] = prompt_flux
print(f"Calculated {len(e_binc_fine)} E_true x {len(t_binc_fine)} Zenith bins for nu and nubar fluxes.")

#Bkg computation: E_true rate
print("Computing atmospheric neutrino rate in E_true (Flux * A_eff)...")
atm_conv_rate_etrue = 0
atm_prompt_rate_etrue = 0
for l,a in zip(['nu', 'nubar'], [aeff_nu_q, aeff_nubar_q]):
    atm_conv_rate_etrue += atm_conv_flux[l] * a
    atm_prompt_rate_etrue += atm_prompt_flux[l] * a
#Background Computation: E_reco Rate (Folding with E_disp)
print("Folding E_true rate with Energy Dispersion (E_disp) to get E_reco rate...")

energy_axis_true = MapAxis.from_edges(e_bins_coarse, unit="GeV", name="energy_true")
energy_axis=MapAxis.from_edges(e_bins_fine_ext, unit='GeV',name='energy')
migra_axis = MapAxis.from_edges(migra_bins, name="migra")
offset_axis = MapAxis.from_edges(t_bins_coarse, unit="deg", name="offset")
axes = MapAxes([energy_axis_true, migra_axis, offset_axis])
data = edisp_q_all.T * sizes_migra_bins[:, None]
edisp_cls = EnergyDispersion2D(axes=axes, data=data)

atm_conv_rate = np.zeros((len(e_binc_fine_ext), len(t_binc_fine)))
atm_prompt_rate = np.zeros((len(e_binc_fine_ext), len(t_binc_fine)))

print(' i  zenith  sum along E_reco')
for i in range(len(t_binc_fine)):
    # Create the Edisp kernel (matrix) for the current zenith angle
    # The kernel is a matrix of size (N_Ereco x N_Etrue)
    edisp_matrix = edisp_cls.to_edisp_kernel(offset=t_binc_fine[i]*u.deg, energy_axis_true=e_bins_fine*u.GeV, energy_axis=e_bins_fine_ext*u.GeV)
    # Convert the E_true rate to total events per bin (Rate * dE)
    conv_etrue = atm_conv_rate_etrue[:,i] * np.diff(e_bins_fine)
    prompt_etrue = atm_prompt_rate_etrue[:,i] * np.diff(e_bins_fine)
    # Fold with the E_disp matrix: E_reco Rate = (E_true Events) * E_disp_Matrix / dE_reco
    # np.dot performs the matrix multiplication: (1 x N_Etrue) * (N_Etrue x N_Ereco) -> (1 x N_Ereco)
    atm_conv_rate[:, i] = np.dot(conv_etrue, edisp_matrix.pdf_matrix) / np.diff(e_bins_fine_ext)
    atm_prompt_rate[:, i] = np.dot(prompt_etrue, edisp_matrix.pdf_matrix) / np.diff(e_bins_fine_ext)

    # Check the sum of the E_disp matrix rows (should be close to 1)
    print('{:>2} {:>6.1f} {:>12.4f}'.format(i, t_binc_fine[i]*180/np.pi, edisp_matrix.pdf_matrix.sum(axis=1).mean()))
```

```python
#-------Calculate Atmospheric muon bkg--------
print("Calculating atmospheric muon background rate...")
df_mu_q = df_mu_q.copy()
df_mu_q['theta'] = zenith_angle(df_mu_q, mc=False)
# Compute the 2D histogram (E_reco x Zenith) of weighted muon events
atm_mu_bkg = np.histogram2d(df_mu_q.E, df_mu_q.theta, bins=(e_bins_fine_ext, t_bins_fine), weights=df_mu_q.weight)[0]
atm_mu_bkg *= 2
atm_mu_rate = atm_mu_bkg / (np.pi * 4 / len(t_binc_fine) * np.diff(e_bins_fine_ext)[:,None])
# ------- Constants & masking ---------
de_domega = 4 * np.pi / len(t_binc_fine) * np.diff(e_bins_fine)
de_domega_reco = 4 * np.pi / len(t_binc_fine) * np.diff(e_bins_fine_ext)
t_mask = t_binc_fine * 180 / np.pi > 80  # compute rate for analysis region
t_year = 365 * 24 * 3600

#--------Smooth background---------
print("Smoothing muon background and preparing output files...")
mu_horizon_raw = atm_mu_rate[:,5]
# apply Gaussian smoothing to the rate along the energy axis
mu_horizon_smoothed = gaussian_filter(mu_horizon_raw, [2.0], mode='nearest')
#define the final output muon background rate array
atm_mu_rate_out = np.zeros_like(atm_mu_rate)
atm_mu_rate_out[:,5] = mu_horizon_smoothed

#write the raw rates
np.savetxt('raw_muon_rate.dat', np.array([e_binc_fine_ext, mu_horizon_raw/1e3]).T, header='E[GeV]  rate[MeV-1 s-1 sr-1]')
np.savetxt('prompt_nu_rate.dat', np.array([e_binc_fine_ext, atm_prompt_rate[:,t_mask].sum(axis=1)/1e3]).T, header='E[GeV]  rate[MeV-1 s-1 sr-1]')

# Define the zenith bin edges for the output muon background file
# This creates bin edges that encompass the analysis region for muons
t_bins_mu_out = np.insert(
    np.arccos(cos_bins_fine[1:] + 1. / len(t_binc_fine)), 0, 0.0)

print("Standard zenith binning (deg):\n", t_bins_fine * 180 / np.pi)
print("Zenith binning for muon bkg file (deg):\n", t_bins_mu_out * 180 / np.pi)

build_background_files(
    e_bins_fine_ext, # E_reco bins
    t_bins_fine, # Zenith bins for neutrinos
    atm_conv_rate, # Conventional E_reco rate
    atm_prompt_rate, # Prompt E_reco rate
    atm_mu_rate_out, # Smoothed muon E_reco rate (only populated in one zenith bin)
    t_bins_mu_out, # Zenith bins for muons
    outdir=".", # Output directory
)
print("Background FITS files saved to '.' directory.")
```

## 9. Plot Atmospheric Background Rates

```python
print("Generating Plot 1: Atmospheric background rates...")
plt.style.use('seaborn-v0_8-whitegrid')
# (Matplotlib settings as in the original script)
plt.rcParams.update({
    'font.size': 7,
    'axes.labelsize': 7,
    'axes.titlesize': 10,
    'legend.fontsize': 6,
    'xtick.labelsize': 8,
    'ytick.labelsize': 8,
    'axes.linewidth': 1,
    'xtick.direction': 'in',
    'ytick.direction': 'in',
    'xtick.major.size': 4,
    'ytick.major.size': 4,
    'xtick.minor.size': 2,
    'ytick.minor.size': 2,
    'legend.frameon': False
})

# Comparison of atmospheric background rates, summed over the analysis zenith bins (t_mask)
fig,ax = plt.subplots(figsize=(3.5, 2.75), dpi=300)

ax.set_title('Atmospheric backgrounds')
# Muon background (E_reco)
ax.plot(e_binc_fine_ext, atm_mu_rate[:,t_mask].sum(axis=1) * de_domega_reco * t_year,
        '+-', label='Muons ($E_\mathrm{reco}$)', ds='steps-mid')
# Conventional Neutrino (E_true)
ax.plot(e_binc_fine, atm_conv_rate_etrue[:,t_mask].sum(axis=1) * de_domega * t_year,
        '+-', label='Conv. neutrinos ($E_\mathrm{true}$)', ds='steps-mid')
# Conventional Neutrino (E_reco) - Folded
ax.plot(e_binc_fine_ext, atm_conv_rate[:,t_mask].sum(axis=1) * de_domega_reco * t_year,
        '+-', label='Conv. neutrinos ($E_\mathrm{reco}$)', ds='steps-mid')
# Prompt Neutrino (E_true)
ax.plot(e_binc_fine, atm_prompt_rate_etrue[:,t_mask].sum(axis=1) * de_domega * t_year,
        '+-', label='Prompt neutrinos ($E_\mathrm{true}$)', ds='steps-mid')
# Prompt Neutrino (E_reco) - Folded
ax.plot(e_binc_fine_ext, atm_prompt_rate[:,t_mask].sum(axis=1) * de_domega_reco * t_year,
        '+-', label='Prompt neutrinos ($E_\mathrm{reco}$)', ds='steps-mid')

ax.loglog()
ax.set_xlabel('$E$ [GeV]')
ax.set_ylabel('Rate [yr$^{-1}$]')
ax.legend(loc='upper right', fontsize='small')
ax.set_ylim(3e-5, 3e4)
ax.grid(True, which='both', ls='--', alpha=0.3)
plt.savefig("plots/atmospheric_rates.png")
plt.show()
print("Plot 1 saved to 'atmospheric_rates.png'.")

```

```python

print("Generating Plot 2: Final background rate comparison (E_reco)...")
fig, ax = plt.subplots(figsize=(3.5, 2.75), dpi=300)

# Conventional Neutrinos (E_reco, summed over t_mask)
ax.plot(e_binc_fine_ext*1e-3, atm_conv_rate[:,t_mask].sum(axis=1) * de_domega_reco * t_year,
        label='Conv. neutrinos')
# Prompt Neutrinos (E_reco, summed over t_mask)
ax.plot(e_binc_fine_ext*1e-3, atm_prompt_rate[:,t_mask].sum(axis=1) * de_domega_reco * t_year,
        label='Prompt neutrinos', ls='--')
# Raw Muons (E_reco, single horizon bin)
ax.plot(e_binc_fine_ext*1e-3, mu_horizon_raw * de_domega_reco * t_year,
        '-', label='Muons (Raw)', ds='steps-mid', zorder=0)
# Smoothed Muons (E_reco, single horizon bin)
ax.plot(e_binc_fine_ext*1e-3, mu_horizon_smoothed * de_domega_reco * t_year,zorder=0,ls='-', label='Muons (Smoothed)')

ax.loglog()
ax.set_xlabel('$E_\mathrm{reco}$ [TeV]') # Energy axis is now in TeV
ax.set_ylabel('Background Rate [yr$^{-1}$]')
ax.legend(loc='upper right', fontsize='small')
ax.set_ylim(3e-5, 3e5)
ax.set_xticks([10**i for i in range(-3,6,2)]) # Ticks in TeV
ax.grid(True, which='both', ls='--', alpha=0.3)
plt.savefig("plots/smoothed_bkg_rates.png")
plt.show()
print("Plot 2 saved to 'smoothed_bkg_rates.png'.")
```
