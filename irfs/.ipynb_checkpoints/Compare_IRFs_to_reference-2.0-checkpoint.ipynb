```python
def read_aeff_with_edges(fits_file):
    """Read AEFF_2D FITS and return bin edges, centers, and values."""
    with fits.open(fits_file) as hdul:
        data = hdul[1].data
        e_lo = data["ENERG_LO"][0]      # length n_energy
        e_hi = data["ENERG_HI"][0]
        theta_lo = data["THETA_LO"][0]  # length n_theta
        theta_hi = data["THETA_HI"][0]
        aeff = data["EFFAREA"][0]       # shape (n_energy, n_theta) or (n_theta, n_energy)

    # Bin centers
    e_center = np.sqrt(e_lo * e_hi)                # geometric mean for energy
    theta_center = 0.5 * (theta_lo + theta_hi)    # arithmetic mean for angle (radians)

    # Bin edges
    e_edges = np.hstack([e_lo, e_hi[-1]])
    theta_edges = np.hstack([theta_lo, theta_hi[-1]])

    return e_edges, e_center, theta_edges, theta_center, aeff
e_edges, e_center, theta_edges, theta_center, aeff_value  = read_aeff_with_edges('../aeff.fits')
print("Shape AEFF:", aeff.data.shape if hasattr(aeff, 'data') else aeff.shape)
print("Numero bin energia:", len(e_center))
print("Numero bin offset/theta:", len(theta_center))
print("Numero bin energia:", len(e_center))
print("Numero bin offset/theta:", len(theta_center))

print("Energy edges:", e_edges)
print("Energy centers:", e_center)
print("Theta edges:", theta_edges)
print("Theta centers:", theta_center)

print("Aeff min:", np.nanmin(aeff_value))
print("Aeff max:", np.nanmax(aeff_value))
print("Aeff mean:", np.nanmean(aeff_value))

print("Primi 5 valori AEFF (energia basso, offset basso):", aeff_value[:5, :5])
print("Ultimi 5 valori AEFF (energia alto, offset alto):", aeff_value[-5:, -5:])
# Mantieni la stessa orientazione di 0.17
# NDDataArray → array NumPy puro
e_edges, e_center, theta_edges, theta_center, aeff_values = read_aeff_with_edges('../aeff.fits')

# Mantieni la stessa orientazione di 0.17
Z = np.nan_to_num(np.log10(aeff_values[:, ::-1]), neginf=-6)
Z[Z <= -2] = -2

print("Shape Z:", Z.shape)  # dovrebbe essere (n_energy, n_zenith)
print("Riga energia più bassa:", Z[0, :])  # confronto con 0.17
print("Riga energia più alta:", Z[-1, :])

# Extract binning
#aeff_e_bins = aeff.data.axis('energy_true').edges.value
#aeff_e_binc = aeff.data.axis('energy_true').center.value
#aeff_zen_bins = aeff.data.axis('offset').edges.value
#aeff_zen_binc = aeff.data.axis('offset').center.value
energy_20 = e_center
aeff_20_values = aeff_values  # shape (n_energy, n_zenith)

# Sum over zenith bins
# Sum over zenith bins (rows)
aeff_20_sum = np.sum(aeff_values, axis=0)  # shape (48,)

# Plot
plt.figure(figsize=(8,6))
plt.plot(energy_20, aeff_20_sum, marker='s', color='C1', label='Gammapy 2.0 / FITS')
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Energy [GeV]")
plt.ylabel("Sum of Effective Area over Zenith [m²]")
plt.title("Gammapy 2.0 / FITS: Effective Area vs Energy (summed over zenith)")
plt.grid(True, which='both', ls='--', alpha=0.5)
plt.legend()

# AEFF 2.0
# theta_center, aeff_values from read_aeff_with_edges()
# aeff_values shape = (n_zenith, n_energy)
energy_idx = 0  # same energy index as above

aeff_vs_theta = aeff_values[:, energy_idx]      # shape (n_zenith,)
theta = theta_center

plt.figure(figsize=(8,6))
plt.plot(theta, aeff_vs_theta, marker='s', color='C1', label=f'Energy = {e_center[energy_idx]:.2f} GeV')
plt.xlabel("Zenith / Offset [rad]")
plt.ylabel("Effective Area [m²]")
plt.title("Gammapy 2.0 / FITS: AEFF vs Zenith")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
from gammapy.irf import PSF3D

# Lettura dei file
psf = PSF3D.read("psf_no_norm.fits")
psf_raw = PSF3D.read("psf_no_smooth_no_norm.fits")

# Estrai gli assi
energy_axis = psf.axes["energy_true"]
offset_axis = psf.axes["offset"]
rad_axis = psf.axes["rad"]

# Bin edges e bin center
psf_e_bins = energy_axis.edges.to_value("GeV")
psf_e_binc = np.sqrt(psf_e_bins[:-1] * psf_e_bins[1:])

psf_rad_bins = rad_axis.edges.to_value("deg")
psf_rad_binc = 0.5 * (psf_rad_bins[:-1] + psf_rad_bins[1:])

# Ottieni i dati della PSF (shape: n_energy, n_offset, n_rad)
psf_data = psf.data  # quantity con unità 1/sr
psf_raw_data = psf_raw.data

# Somma su offset
psf_zsum = psf_data.sum(axis=1)       # (24, 111)
psf_raw_zsum = psf_raw_data.sum(axis=1)

# Correzione per la larghezza dei bin radiali
bw = np.diff(psf_rad_bins ** 2) * (np.pi / 180) ** 2 * np.pi  # steradianti

# ✅ Broadcasting corretto
psf_zsum_scl = psf_zsum * bw[None, :]
psf_raw_zsum_scl = psf_raw_zsum * bw[None, :]

# --- Verifica / confronto (Gammapy 2.0) ---
import numpy as np

print("Gammapy 2.0: statistiche PSF integrate")
print("---------------------------------------")
print(f"Shape: {psf_zsum_scl.shape}")
print(f"Valore medio : {np.mean(psf_zsum_scl):.3e}")
print(f"Valore max   : {np.max(psf_zsum_scl):.3e}")
print(f"Valore min   : {np.min(psf_zsum_scl):.3e}")

```

```python
def find_psf_quantiles(rad_values, psf_data, quantiles):
    # cumsum and normalize
    cumsum = np.cumsum(psf_data, axis=1).astype(float)
    valid = cumsum[:, -1] > 0
    cumsum[valid] /= cumsum[valid][:, -1][:, np.newaxis]

    # preallocate results
    ang_res = np.zeros((len(quantiles), psf_data.shape[0]))

    for i, q in enumerate(quantiles):
        # for each energy slice
        for j in range(psf_data.shape[0]):
            ang_res[i, j] = np.interp(q, cumsum[j], rad_values)

    return ang_res

```

```python
# Find quantiles for reference PSF
km3_ang_res_nu = find_psf_quantiles(km3_psf_rad_binc, km3_psf_nu_data_zsum, psf_quantiles)
km3_ang_res_nubar = find_psf_quantiles(km3_psf_rad_binc, km3_psf_nubar_data_zsum, psf_quantiles)

# Average over nu/nubar
km3_ang_res = []
for i in range(len(psf_quantiles)):
    km3_ang_res.append(0.5 * (km3_ang_res_nu[i] + km3_ang_res_nubar[i]))
```

```python
# Load energy dispersion from reference file
km3_edisp_nu = ref_file['h_EREC_numuCC_track;1']
km3_edisp_nubar = ref_file['h_EREC_anumuCC_track;1']

# extract data and binning
km3_edisp_nu_data, km3_edisp_nu_loge_bins, km3_edisp_nu_theta_bins, km3_edisp_nu_logereco_bins = km3_edisp_nu.to_numpy()
km3_edisp_nubar_data, km3_edisp_nubar_loge_bins, km3_edisp_nubar_theta_bins, km3_edisp_nubar_logereco_bins = km3_edisp_nubar.to_numpy()

assert (km3_edisp_nu_loge_bins == km3_edisp_nubar_loge_bins).all()
assert (km3_edisp_nu_theta_bins == km3_edisp_nubar_theta_bins).all()
assert (km3_edisp_nu_logereco_bins == km3_edisp_nubar_logereco_bins).all()

km3_edisp_e_bins = 10 ** km3_edisp_nu_loge_bins
km3_edisp_ereco_bins = 10 ** km3_edisp_nu_logereco_bins

km3_edisp_e_binc = np.sqrt(km3_edisp_e_bins[:-1] * km3_edisp_e_bins[1:])
km3_edisp_ereco_binc = np.sqrt(km3_edisp_ereco_bins[:-1] * km3_edisp_ereco_bins[1:])

# sum over zenith angle axis
km3_edisp_nu_data_zsum = km3_edisp_nu_data.sum(axis=1)
km3_edisp_nubar_data_zsum = km3_edisp_nubar_data.sum(axis=1)

# Determine valid energy range
km3_edisp_valid = (km3_edisp_nu_data_zsum.sum(axis=1) > 0) & (km3_edisp_nubar_data_zsum.sum(axis=1) > 0)
```

```python
from gammapy.irf import EnergyDispersion2D
import numpy as np

# --- Leggi EDisp IRF (unnormalised + raw) ---
edisp = EnergyDispersion2D.read("edisp_no_norm.fits", hdu="EDISP_2D")
edisp_raw = EnergyDispersion2D.read("edisp_no_smooth_no_norm.fits", hdu="EDISP_2D")

# --- Assi ---
energy = edisp.axes["energy_true"]   # GeV
migra  = edisp.axes["migra"]         # dimensionless
theta  = edisp.axes["offset"]        # deg, opzionale

# --- Costruisci bin centrali ---
edisp_e_bins = np.append(energy.edges.value[:-1], energy.edges.value[-1])
edisp_e_binc = np.sqrt(edisp_e_bins[:-1] * edisp_e_bins[1:])

edisp_migra_bins = np.append(migra.edges.value[:-1], migra.edges.value[-1])
edisp_migra_binc = np.sqrt(edisp_migra_bins[:-1] * edisp_migra_bins[1:])

# --- Somma su offset/zenith ---
edisp_zsum = np.array(edisp.data.data).sum(axis=2)       # shape: (n_energy, n_migra)
edisp_raw_zsum = np.array(edisp_raw.data.data).sum(axis=2)

# --- Scaling per bin width (opzionale) ---
bw = np.diff(edisp_migra_bins)
edisp_zsum_scl = edisp_zsum * bw[None, :]
edisp_raw_zsum_scl = edisp_raw_zsum * bw[None, :]




```

```python
# Find quantiles for reference EDISP
km3_e_res_nu = find_psf_quantiles(km3_edisp_ereco_binc, km3_edisp_nu_data_zsum, edisp_quantiles)
km3_e_res_nubar = find_psf_quantiles(km3_edisp_ereco_binc, km3_edisp_nubar_data_zsum, edisp_quantiles)

# Average over nu/nubar
km3_e_res = []
for i in range(len(edisp_quantiles)):
    km3_e_res.append(0.5 * (km3_e_res_nu[i] + km3_e_res_nubar[i]))
```

```python
# Compute quantiles for our EDISP
e_res = find_psf_quantiles(edisp_migra_binc, edisp_zsum, edisp_quantiles)
e_res_raw = find_psf_quantiles(edisp_migra_binc, edisp_raw_zsum, edisp_quantiles)
```
