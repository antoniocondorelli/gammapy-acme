## 7. Point Spread Function (PSF)

Same idea as Edisp: either build it from event-level MC, or convert an external PSF table to a GADF PSF file and set `PSF_FITS`.


```python
# Optional: set this if you already have a PSF file in GADF format
PSF_FITS = "psf_no_zenith.fits"  # e.g. "psf_no_zenith.fits"

print("PSF_FITS =", PSF_FITS)
```

```python
# Optional quick check (only if you provided PSF_FITS)
if PSF_FITS is not None:
    import astropy.units as u
    from gammapy.irf import PSF3D

    psf_all = PSF3D.read(PSF_FITS)
    print(psf_all)
else:
    print("Skipping PSF checks (no PSF_FITS provided).")
```

```python
# ------------------------------------------------------------------------------
# 3b) Containment radii (68% and 95%) vs energy
# ------------------------------------------------------------------------------

# Get the true-energy axis from the PSF object
energy_axis = psf_all.axes["energy_true"]
energies = energy_axis.center.to("TeV")

# Radial grid (same as above, but in radians for integration)
rad_grid = np.linspace(0.0, 5.0, 200) * u.deg
rad_grid_rad = rad_grid.to("rad").value

# Small step in radius (r_{i+1} - r_i) in radians
dr = np.diff(rad_grid_rad)
rad_mid = 0.5 * (rad_grid_rad[:-1] + rad_grid_rad[1:])

# Solid angle element for a thin ring at radius r (exact on the sphere):
# dΩ ≈ 2π sin(r) dr
dOmega = 2.0 * np.pi * np.sin(rad_mid) * dr  # [sr]

r68_list = []
r95_list = []

for E in energies:
    # Evaluate PSF on the radial grid for this energy
    vals = psf_all.evaluate(
        rad=rad_grid,
        energy_true=E,
        offset=offset,
    )  # sr^-1

    # Use mid-bin values in the integral
    vals_mid = vals.to_value("1/sr")[:-1]

    # Differential probability in each ring: dP = PSF(r) * dΩ
    dP = vals_mid * dOmega  # dimensionless

    # Cumulative distribution as a function of radius
    cdf = np.cumsum(dP)
    if cdf[-1] <= 0:
        # No PSF normalization for this energy; skip safely
        r68_list.append(np.nan)
        r95_list.append(np.nan)
        continue

    cdf /= cdf[-1]

    # Interpolate to find radii where CDF reaches 68% and 95%
    # We restrict to strictly increasing region to avoid numerical edge cases
    mask_pos = cdf > 0
    cdf_valid = cdf[mask_pos]
    rad_valid = rad_mid[mask_pos]

    r68 = np.interp(0.68, cdf_valid, rad_valid)
    r95 = np.interp(0.95, cdf_valid, rad_valid)

    # Convert back to degrees
    r68_list.append(r68 * 180.0 / np.pi)
    r95_list.append(r95 * 180.0 / np.pi)

r68_arr = np.array(r68_list)
r95_arr = np.array(r95_list)

fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(energies.to_value("TeV"), r68_arr, label="68% containment")
ax.plot(energies.to_value("TeV"), r95_arr, label="95% containment")

ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("Energy [TeV]")
ax.set_ylabel("Containment radius [deg]")
ax.set_title("PSF containment radii vs energy")
ax.legend()
plt.tight_layout()
plt.show()
```

# Energy dispersion (EDISP)

In this section we build a **GADF** `EDISP_2D` FITS file starting from a CSV that contains an energy-response matrix.

**Assumptions about your CSV**
- columns `log10(nu_E [GeV]) low/center/high`
- columns `log10(reco_E [GeV]) low/center/high`
- a column labeled `dP/dlog10(nu_E [GeV])` in your example; since for each true-energy bin you sweep reconstructed-energy bins, the code interprets it as a **density** `dP/dlog10(E_reco)`.

The code below:
- builds true-energy edges (GeV)
- builds the **MIGRA = E_reco / E_true** axis
- converts the density to per-bin probabilities and normalizes for each true-energy bin
- writes a GADF FITS with a single **THETA** bin (0..π), like for AEFF.

# Point Spread Function (PSF)

In this section we build a **GADF** `PSF_2D_TABLE` FITS file starting from a CSV that contains:
- `log10(nu_E [GeV]) low/center/high`
- `log10(psi [degrees])` (sample points in log10 of the angular separation)
- `dP/dOmega` (density per steradian, typically `sr^-1`)

The code:
- reconstructs **RAD** bin edges (in degrees) from the sampled points
- fills `RPSF` per bin
- writes a GADF FITS with a single **THETA** bin.

```python
from gammapy.irf import Background2D
import astropy.units as u
import numpy as np
from scipy.ndimage import gaussian_filter1d

# Read muon background FITS
bkg2 = Background2D.read("bkg_total.fits")

# Extract axes
e_edges = bkg2.axes["energy"].edges.to_value("GeV")
dec_edges_deg = bkg2.axes["offset"].edges.to_value("deg")

# Convert declination edges back to sin(dec)
dec_edges_rad = np.deg2rad(dec_edges_deg)
sin_edges = np.sin(dec_edges_rad)

# Extract data (remove units)
# Extract data safely (works whether data is Quantity or ndarray)
data_obj = bkg2.data

if hasattr(data_obj, "to_value"):
    mu_rate_diff = data_obj.to_value("s-1 MeV-1 sr-1")
else:
    mu_rate_diff = np.asarray(data_obj)

# Energy bin widths
dE_GeV = np.diff(e_edges)
dE_MeV = dE_GeV * 1e3

# Solid angle per declination band:
# dΩ = 2π (sin_hi - sin_lo)
dOmega = 2.0 * np.pi * np.diff(sin_edges)

# Convert differential rate -> rate per bin
mu_rate_bin = mu_rate_diff * dE_MeV[:, None] * dOmega[None, :]  # [s^-1]
```

```python
# Choose declination bin index (example: near horizon)
horizon_index = len(dec_edges_deg) // 2

mu_horizon_raw = mu_rate_bin[:, horizon_index]
import numpy as np
from scipy.ndimage import gaussian_filter1d

def smooth_in_logE(y, e_centers_GeV, sigma_decades=0.15, eps=1e-40):
    """
    Smooth y(logE) assuming e_centers are logarithmically spaced or near-log spaced.
    sigma_decades = width in log10(E) (e.g. 0.10–0.25 decades).
    """
    y = np.asarray(y, float)
    e = np.asarray(e_centers_GeV, float)

    logE = np.log10(e)
    dlogE = np.median(np.diff(logE))
    if dlogE <= 0:
        raise ValueError("Energy centers must be strictly increasing")

    sigma_bins = sigma_decades / dlogE

    # smooth in log-space of y to preserve power-laws
    y_clip = np.clip(y, eps, None)
    logy = np.log10(y_clip)
    logy_sm = gaussian_filter1d(logy, sigma=sigma_bins, mode="nearest")
    return 10**logy_sm

mu_horizon_smoothed = smooth_in_logE(mu_horizon_raw, e_binc_fine_ext, sigma_decades=0.20)

```

```python
import matplotlib.pyplot as plt

print("Generating final background comparison plot...")

t_year = 365 * 24 * 3600

e_centers = np.sqrt(e_edges[:-1] * e_edges[1:])
E_reco_TeV = e_centers * 1e-3

# --- neutrinos ---
theta_centers = 0.5 * (theta_edges[:-1] + theta_edges[1:])
t_mask = (theta_centers * 180/np.pi) > 80

dOmega_theta = 2*np.pi*(np.cos(theta_edges[:-1]) - np.cos(theta_edges[1:]))

conv_year = (
    atm_conv_rate[:, t_mask]
    * dOmega_theta[t_mask][None, :]
    * dE_MeV[:, None]
).sum(axis=1) * t_year

prompt_year = (
    atm_prompt_rate[:, t_mask]
    * dOmega_theta[t_mask][None, :]
    * dE_MeV[:, None]
).sum(axis=1) * t_year

# --- muons ---
mu_raw_year = mu_horizon_raw * t_year
mu_smooth_year = mu_horizon_smoothed * t_year

# --- Plot ---
fig, ax = plt.subplots(figsize=(3.5, 2.75), dpi=300)

ax.plot(E_reco_TeV, conv_year, label="Conventional neutrinos")
ax.plot(E_reco_TeV, prompt_year, label="Prompt neutrinos", ls="--")
ax.plot(E_reco_TeV, mu_raw_year, label="Muons (raw)", drawstyle="steps-mid")
ax.plot(E_reco_TeV, mu_smooth_year, label="Muons (smoothed)")

ax.loglog()
ax.set_xlabel(r"$E_\mathrm{reco}$ [TeV]")
ax.set_ylabel(r"Background rate [yr$^{-1}$]")
ax.legend()
ax.grid(True, which="both", ls="--", alpha=0.3)

plt.tight_layout()
plt.show()
```
