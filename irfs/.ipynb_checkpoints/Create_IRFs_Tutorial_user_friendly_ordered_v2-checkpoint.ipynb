# Building Instrument Response Functions (IRFs) for KM3NeT

This notebook guides you through the process of building IRFs for a KM3NeT analysis.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import astropy.units as u
from gammapy.irf import EffectiveAreaTable2D, EnergyDispersion2D, PSF3D

from scipy.ndimage import gaussian_filter1d, gaussian_filter

from helpers import (
    build_aeff_from_table_no_zenith,
    build_edisp_from_csv,
    build_psf_from_csv,
    build_background2d_from_csv_v2,
    build_atmospheric_background_from_irfs,
)

# A single offset value is enough here because we build zenith-independent IRFs
offset = 0.5 * u.deg
```

## Data paths
Edit these paths to point to your local files.


```python
AEFF_TABLE_PATH = "data/aeff_coszen_allnumuCC_track.csv"
EDISP_CSV_PATH = "data/energyresponse_numuCC_track.csv"
PSF_CSV_PATH   = "data/psf_numuCC_track.csv"
BKG_CSV_PATH   = "data/bkg_track.csv"

AEFF_FITS  = "aeff_no_zenith.fits"
EDISP_FITS = "edisp_no_zenith.fits"
PSF_FITS   = "psf_no_zenith.fits"
BKG_FITS   = "bkg_total.fits"
```

## 1) Effective area (Aeff)
We convert an input table into a GADF `AEFF_2D` FITS file, then we sanity-check the result.


```python
aeff, e_bins, t_bins = build_aeff_from_table_no_zenith(
    AEFF_TABLE_PATH,
    AEFF_FITS,
    zenith_reduce="mean",  # "mean" | "median"
    upgoing_only=False,    # set True if your analysis is strictly up-going
)

print(f"Wrote: {AEFF_FITS}")
```

### Quick look
Plot Aeff vs true energy.


```python
aeff_irf = EffectiveAreaTable2D.read(AEFF_FITS)

fig, ax = plt.subplots(figsize=(7, 5))
try:
    aeff_1d = aeff_irf.to_effective_area_table(offset=offset)
    aeff_1d.plot(ax=ax)
except Exception:
    energies = aeff_irf.axes["energy_true"].center.to("TeV")
    vals = aeff_irf.evaluate(energy_true=energies, offset=offset).to("m2")
    ax.loglog(energies.value, vals.value)

ax.set_xlabel("True energy [TeV]")
ax.set_ylabel(r"A$_\mathrm{eff}$ [m$^2$]")
ax.grid(True, which="both", alpha=0.3)
plt.show()
```

## 2) Energy dispersion (Edisp)
We convert a CSV table into a GADF `EDISP_2D` FITS file.


```python
matrix_edisp, e_true_edges, migra_edges, theta_edges = build_edisp_from_csv(
    EDISP_CSV_PATH,
    EDISP_FITS,
)

print(f"Wrote: {EDISP_FITS}")
```

### Quick check
Read back the file and inspect the object.


```python
edisp = EnergyDispersion2D.read(EDISP_FITS)
print(edisp)

# Example: show the migration distribution at one true-energy bin
e_true = edisp.axes["energy_true"].center[edisp.axes["energy_true"].nbin // 2]
migra = edisp.axes["migra"].center

vals = edisp.evaluate(energy_true=e_true, migra=migra, offset=offset).to_value("")
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(migra, vals)
ax.set_xlabel("Migra = E_reco / E_true")
ax.set_ylabel("Probability density")
ax.set_title(f"Migration distribution at E_true = {e_true.to('TeV'):.2g}")
ax.grid(True, alpha=0.3)
plt.show()
```

### Student exercises
1. Pick a different `e_true` (low / mid / high energy) and compare the migration distributions.
2. Compute the **bias** and **resolution** at your chosen `e_true`:
   - bias = mean(migra) - 1
   - resolution = std(migra)
3. Optional: overlay two curves (low vs high energy) on the same plot.


```python
# TODO (students): change the index to probe different energies
idx = edisp.axes["energy_true"].nbin // 2
e_true = edisp.axes["energy_true"].center[idx]
migra = edisp.axes["migra"].center

pdf = edisp.evaluate(energy_true=e_true, migra=migra, offset=offset).to_value("")

# Normalize numerically (robust if binning is non-uniform)
dm = np.gradient(migra)
norm = np.sum(pdf * dm)
pdfn = pdf / norm if norm > 0 else pdf

mean = np.sum(migra * pdfn * dm)
var  = np.sum((migra - mean) ** 2 * pdfn * dm)

bias = mean - 1.0
res  = np.sqrt(var)

print(f"E_true = {e_true.to('TeV'):.3g} | bias = {bias:.4f} | resolution = {res:.4f}")
```

## 3) Point Spread Function (PSF)
We convert a CSV table into a GADF `PSF_2D_TABLE` FITS file and compute containment radii.


```python
rpsf, e_true_edges_psf, rad_edges, theta_edges_psf = build_psf_from_csv(
    PSF_CSV_PATH,
    PSF_FITS,
)

psf_all = PSF3D.read(PSF_FITS)
print(psf_all)
print(f"Wrote: {PSF_FITS}")
```

### Containment radii (68% and 95%) vs energy


```python
# ------------------------------------------------------------------------------
# 3b) Containment radii (68% and 95%) vs energy
# ------------------------------------------------------------------------------

# Get the true-energy axis from the PSF object
energy_axis = psf_all.axes["energy_true"]
energies = energy_axis.center.to("TeV")

# Radial grid (same as above, but in radians for integration)
rad_grid = np.linspace(0.0, 5.0, 200) * u.deg
rad_grid_rad = rad_grid.to("rad").value

# Small step in radius (r_{i+1} - r_i) in radians
dr = np.diff(rad_grid_rad)
rad_mid = 0.5 * (rad_grid_rad[:-1] + rad_grid_rad[1:])

# Solid angle element for a thin ring at radius r (exact on the sphere):
# dΩ ≈ 2π sin(r) dr
dOmega = 2.0 * np.pi * np.sin(rad_mid) * dr  # [sr]

r68_list = []
r95_list = []

for E in energies:
    # Evaluate PSF on the radial grid for this energy
    vals = psf_all.evaluate(
        rad=rad_grid,
        energy_true=E,
        offset=offset,
    )  # sr^-1

    # Use mid-bin values in the integral
    vals_mid = vals.to_value("1/sr")[:-1]

    # Differential probability in each ring: dP = PSF(r) * dΩ
    dP = vals_mid * dOmega  # dimensionless

    # Cumulative distribution as a function of radius
    cdf = np.cumsum(dP)
    if cdf[-1] <= 0:
        # No PSF normalization for this energy; skip safely
        r68_list.append(np.nan)
        r95_list.append(np.nan)
        continue

    cdf /= cdf[-1]

    # Interpolate to find radii where CDF reaches 68% and 95%
    # We restrict to strictly increasing region to avoid numerical edge cases
    mask_pos = cdf > 0
    cdf_valid = cdf[mask_pos]
    rad_valid = rad_mid[mask_pos]

    r68 = np.interp(0.68, cdf_valid, rad_valid)
    r95 = np.interp(0.95, cdf_valid, rad_valid)

    # Convert back to degrees
    r68_list.append(r68 * 180.0 / np.pi)
    r95_list.append(r95 * 180.0 / np.pi)

r68_arr = np.array(r68_list)
r95_arr = np.array(r95_list)

fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(energies.to_value("TeV"), r68_arr, label="68% containment")
ax.plot(energies.to_value("TeV"), r95_arr, label="95% containment")

ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("Energy [TeV]")
ax.set_ylabel("Containment radius [deg]")
ax.set_title("PSF containment radii vs energy")
ax.legend()
plt.tight_layout()
plt.show()
```

## 4) Background (optional)

If you have a background table (typically from simulations), you can convert it to a GADF-compatible `BACKGROUND_2D` IRF.

**Goal:** produce a FITS file that can be used together with the other IRFs (AEFF / EDISP / PSF).

### Student questions
1. What happens to the background rate if you change the energy binning (coarser vs finer)?
2. Try evaluating the background at a different `offset` (e.g. 0.5°, 1.0°, 2.0°). How does it change?


```python
# Build BACKGROUND_2D from CSV
from gammapy.irf import Background2D

bkg, e_edges, fov_edges = build_background2d_from_csv_v2(
    BKG_CSV_PATH,
    out_fits=BKG_FITS,
)

print(f"Wrote: {BKG_FITS}")
print(bkg)
```

```python
# Quick sanity check: evaluate the background rate vs reconstructed energy at one offset
bkg2d = Background2D.read(BKG_FITS)

ereco = bkg2d.axes["energy"].center
offsets = bkg2d.axes["offset"].center

# TODO (students): change which offset bin you use
ioff = bkg2d.axes["offset"].nbin // 2
off = offsets[ioff]

rate = bkg2d.evaluate(energy=ereco, offset=off).to(1 / (u.s * u.MeV * u.sr))

fig, ax = plt.subplots(figsize=(7, 4))
ax.loglog(ereco.to_value("MeV"), rate.to_value(1 / (u.s * u.MeV * u.sr)))
ax.set_xlabel("Reconstructed energy [MeV]")
ax.set_ylabel(r"Background rate [1 / (s MeV sr)]")
ax.set_title(f"Background rate at offset = {off.to('deg'):.2g}")
ax.grid(True, which="both", alpha=0.3)
plt.show()
```
