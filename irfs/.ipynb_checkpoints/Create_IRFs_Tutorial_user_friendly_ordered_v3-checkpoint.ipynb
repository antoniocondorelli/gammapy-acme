## 3) Point Spread Function (PSF)
We convert a CSV table into a GADF `PSF_2D_TABLE` FITS file and compute containment radii.


```python
rpsf, e_true_edges_psf, rad_edges, theta_edges_psf = build_psf_from_csv(
    PSF_CSV_PATH,
    PSF_FITS,
)

psf_all = PSF3D.read(PSF_FITS)
print(psf_all)
print(f"Wrote: {PSF_FITS}")
```

### Containment radii (68% and 95%) vs energy


```python
# ------------------------------------------------------------------------------
# 3b) Containment radii (68% and 95%) vs energy
# ------------------------------------------------------------------------------

# Get the true-energy axis from the PSF object
energy_axis = psf_all.axes["energy_true"]
energies = energy_axis.center.to("TeV")

# Radial grid (same as above, but in radians for integration)
rad_grid = np.linspace(0.0, 5.0, 200) * u.deg
rad_grid_rad = rad_grid.to("rad").value

# Small step in radius (r_{i+1} - r_i) in radians
dr = np.diff(rad_grid_rad)
rad_mid = 0.5 * (rad_grid_rad[:-1] + rad_grid_rad[1:])

# Solid angle element for a thin ring at radius r (exact on the sphere):
# dΩ ≈ 2π sin(r) dr
dOmega = 2.0 * np.pi * np.sin(rad_mid) * dr  # [sr]

r68_list = []
r95_list = []

for E in energies:
    # Evaluate PSF on the radial grid for this energy
    vals = psf_all.evaluate(
        rad=rad_grid,
        energy_true=E,
        offset=offset,
    )  # sr^-1

    # Use mid-bin values in the integral
    vals_mid = vals.to_value("1/sr")[:-1]

    # Differential probability in each ring: dP = PSF(r) * dΩ
    dP = vals_mid * dOmega  # dimensionless

    # Cumulative distribution as a function of radius
    cdf = np.cumsum(dP)
    if cdf[-1] <= 0:
        # No PSF normalization for this energy; skip safely
        r68_list.append(np.nan)
        r95_list.append(np.nan)
        continue

    cdf /= cdf[-1]

    # Interpolate to find radii where CDF reaches 68% and 95%
    # We restrict to strictly increasing region to avoid numerical edge cases
    mask_pos = cdf > 0
    cdf_valid = cdf[mask_pos]
    rad_valid = rad_mid[mask_pos]

    r68 = np.interp(0.68, cdf_valid, rad_valid)
    r95 = np.interp(0.95, cdf_valid, rad_valid)

    # Convert back to degrees
    r68_list.append(r68 * 180.0 / np.pi)
    r95_list.append(r95 * 180.0 / np.pi)

r68_arr = np.array(r68_list)
r95_arr = np.array(r95_list)

fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(energies.to_value("TeV"), r68_arr, label="68% containment")
ax.plot(energies.to_value("TeV"), r95_arr, label="95% containment")

ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("Energy [TeV]")
ax.set_ylabel("Containment radius [deg]")
ax.set_title("PSF containment radii vs energy")
ax.legend()
plt.tight_layout()
plt.show()
```

This step needs **Aeff + Edisp** and an assumed flux model.

```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
from gammapy.irf import Background2D, EffectiveAreaTable2D, EnergyDispersion2D
import importlib
import helpers

importlib.reload(helpers)
# -----------------------
# Inputs
# -----------------------
edisp_csv  = "data/energyresponse_numuCC_track.csv"
edisp_fits = "edisp_no_zenith.fits"
aeff_fits  = "aeff_no_zenith.fits"

print("Building atmospheric background from IRFs...")

# -----------------------
# 1) Build + read EDISP (FITS is the source of truth for binning)
# -----------------------
edisp_out = build_edisp_from_csv(edisp_csv, edisp_fits)
matrix_edisp = edisp_out  # builder will extract [0] if edisp_out is a tuple/list

edisp_irf = EnergyDispersion2D.read(edisp_fits)

e_true_edges = edisp_irf.axes["energy_true"].edges.to_value(u.GeV)
migra_edges  = edisp_irf.axes["migra"].edges.to_value("")  # dimensionless
theta_edges  = edisp_irf.axes["offset"].edges.to_value(u.rad)

# -----------------------
# 2) Read AEFF
# -----------------------
aeff_irf = EffectiveAreaTable2D.read(aeff_fits)

# -----------------------
# 3) Build atmospheric background (conv + prompt)
# -----------------------
atm_conv_rate, atm_prompt_rate, e_bins_reco, theta_edges_out, theta_unit = (
    build_atmospheric_background_from_irfs(
        aeff_irf=aeff_irf,
        edisp_q_all=matrix_edisp,
        e_true_edges_coarse=e_true_edges,
        migra_edges=migra_edges,
        theta_edges_coarse=theta_edges,
    )
)

E_reco_centers_GeV = np.sqrt(e_bins_reco[:-1] * e_bins_reco[1:])

conv_1d   = atm_conv_rate.sum(axis=1)
prompt_1d = atm_prompt_rate.sum(axis=1)

plt.figure(figsize=(4, 3), dpi=150)
plt.loglog(E_reco_centers_GeV, conv_1d, label="Conventional")
plt.loglog(E_reco_centers_GeV, prompt_1d, label="Prompt", ls="--")

plt.xlabel(r"$E_\mathrm{reco}$ [GeV]")
plt.ylabel(r"Rate [s$^{-1}$ GeV$^{-1}$ sr$^{-1}$] (summed over $\theta$ bins)")
plt.grid(True, which="both", ls="--", alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

```
