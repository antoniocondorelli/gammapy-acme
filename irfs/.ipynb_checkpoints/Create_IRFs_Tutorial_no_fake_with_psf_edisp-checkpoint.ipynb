# Building Instrument Response Functions (IRFs) for KM3NeT

This notebook guides you through the process of building IRFs for a KM3NeT analysis.

```python
import sys
import numpy as np
from scipy.ndimage.filters import gaussian_filter1d, gaussian_filter
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

from utils import *
from irf_helper import *
from bkg_functions import build_background_files
```

## 1. File Setup and Data Loading

```python
import numpy as np
import pandas as pd
from astropy.io import fits

# ---------------------------------------------------------------------
# Load your external Aeff table and convert it to a Gammapy/GADF AEFF_2D
# with *no zenith dependence* (single theta bin).
# ---------------------------------------------------------------------

AEFF_TABLE_PATH = "aeff_table.csv"   # <-- set this to your file
AEFF_FITS_OUT   = "aeff_no_zenith.fits"

def build_aeff_from_table_no_zenith(
    aeff_table_path: str,
    file_name: str,
    zenith_reduce: str = "mean",   # "mean" | "median"
    upgoing_only: bool = False,    # True -> only rows with cos(zen)_center < 0
):
    df = pd.read_csv(aeff_table_path)

    # Map the column names shown in your example to short internal names
    col_map = {
        "log10(nu_E [GeV]) low": "log10E_lo",
        "log10(nu_E [GeV]) center": "log10E_c",
        "log10(nu_E [GeV]) high": "log10E_hi",
        "cos(zen) low": "cosz_lo",
        "cos(zen) center": "cosz_c",
        "cos(zen) high": "cosz_hi",
        "aeff [m^2]": "aeff_m2",
    }
    df = df.rename(columns={k: v for k, v in col_map.items() if k in df.columns})

    required = ["log10E_lo", "log10E_hi", "cosz_c", "aeff_m2"]
    missing = [c for c in required if c not in df.columns]
    if missing:
        raise ValueError(f"Missing columns: {missing}. Found columns: {list(df.columns)}")

    if upgoing_only:
        df = df[df["cosz_c"] < 0].copy()

    # Energy bin edges from the table (in log10 GeV)
    loge_lo = np.sort(df["log10E_lo"].unique())
    loge_hi = np.sort(df["log10E_hi"].unique())
    if len(loge_lo) == len(loge_hi):
        edges_log10 = np.concatenate([[loge_lo.min()], loge_hi])
    else:
        edges_log10 = np.sort(np.unique(np.concatenate([df["log10E_lo"].values, df["log10E_hi"].values])))

    e_bins = 10.0 ** edges_log10  # GeV
    e_binc = np.sqrt(e_bins[:-1] * e_bins[1:])

    # Reduce over zenith bins for each energy bin
    gb = df.groupby(["log10E_lo", "log10E_hi"], sort=True)["aeff_m2"]
    if zenith_reduce == "mean":
        aeff_1d = gb.mean().to_numpy()
    elif zenith_reduce == "median":
        aeff_1d = gb.median().to_numpy()
    else:
        raise ValueError("zenith_reduce must be 'mean' or 'median'.")

    # Build 2D array with a single theta bin: shape (nE, 1)
    aeff = aeff_1d[:, None]

    # Single theta bin covering the full sphere (no zenith dependence)
    t_bins = np.array([0.0, np.pi], dtype=float)

    # Write GADF AEFF_2D
    hdu = fits.PrimaryHDU()
    col1 = fits.Column(name="ENERG_LO", format=f"{len(e_binc)}E", unit="GeV", array=[e_bins[:-1]])
    col2 = fits.Column(name="ENERG_HI", format=f"{len(e_binc)}E", unit="GeV", array=[e_bins[1:]])
    col3 = fits.Column(name="THETA_LO", format="1E", unit="rad", array=[t_bins[:-1]])
    col4 = fits.Column(name="THETA_HI", format="1E", unit="rad", array=[t_bins[1:]])
    col5 = fits.Column(
        name="EFFAREA",
        format=f"{len(e_binc) * 1}D",
        dim=f"({len(e_binc)},1)",
        unit="m2",
        array=[aeff.T],
    )
    cols = fits.ColDefs([col1, col2, col3, col4, col5])
    hdu2 = fits.BinTableHDU.from_columns(cols)

    hdu2.header["EXTNAME"] = "EFFECTIVE AREA"
    hdu2.header["HDUDOC"] = "https://github.com/open-gamma-ray-astro/gamma-astro-data-formats"
    hdu2.header["HDUVERS"] = "0.2"
    hdu2.header["HDUCLASS"] = "GADF"
    hdu2.header["HDUCLAS1"] = "RESPONSE"
    hdu2.header["HDUCLAS2"] = "EFF_AREA"
    hdu2.header["HDUCLAS3"] = "FULL-ENCLOSURE"
    hdu2.header["HDUCLAS4"] = "AEFF_2D"

    fits.HDUList([hdu, hdu2]).writeto(file_name, overwrite=True)
    print(f"Wrote: {file_name}")

    return aeff, e_bins, t_bins

aeff, e_bins, t_bins = build_aeff_from_table_no_zenith(
    AEFF_TABLE_PATH,
    AEFF_FITS_OUT,
    zenith_reduce="mean",
    upgoing_only=False,  # set True if your analysis is strictly up-going
)
```

## 7. Point Spread Function (PSF)

Same idea as Edisp: either build it from event-level MC, or convert an external PSF table to a GADF PSF file and set `PSF_FITS`.


```python
# Optional: set this if you already have a PSF file in GADF format
PSF_FITS = None  # e.g. "psf_no_zenith.fits"

print("PSF_FITS =", PSF_FITS)
```

```python
# Optional quick check (only if you provided PSF_FITS)
if PSF_FITS is not None:
    import astropy.units as u
    from gammapy.irf import PSF3D

    psf_all = PSF3D.read(PSF_FITS)
    print(psf_all)
else:
    print("Skipping PSF checks (no PSF_FITS provided).")
```

```python
# ------------------------------------------------------------------------------
# 3b) Containment radii (68% and 95%) vs energy
# ------------------------------------------------------------------------------

# Get the true-energy axis from the PSF object
energy_axis = psf_all.axes["energy_true"]
energies = energy_axis.center.to("TeV")

# Radial grid (same as above, but in radians for integration)
rad_grid = np.linspace(0.0, 5.0, 200) * u.deg
rad_grid_rad = rad_grid.to("rad").value

# Small step in radius (r_{i+1} - r_i) in radians
dr = np.diff(rad_grid_rad)
rad_mid = 0.5 * (rad_grid_rad[:-1] + rad_grid_rad[1:])

# Solid angle element for a thin ring at radius r (exact on the sphere):
# dΩ ≈ 2π sin(r) dr
dOmega = 2.0 * np.pi * np.sin(rad_mid) * dr  # [sr]

r68_list = []
r95_list = []

for E in energies:
    # Evaluate PSF on the radial grid for this energy
    vals = psf_all.evaluate(
        rad=rad_grid,
        energy_true=E,
        offset=offset,
    )  # sr^-1

    # Use mid-bin values in the integral
    vals_mid = vals.to_value("1/sr")[:-1]

    # Differential probability in each ring: dP = PSF(r) * dΩ
    dP = vals_mid * dOmega  # dimensionless

    # Cumulative distribution as a function of radius
    cdf = np.cumsum(dP)
    if cdf[-1] <= 0:
        # No PSF normalization for this energy; skip safely
        r68_list.append(np.nan)
        r95_list.append(np.nan)
        continue

    cdf /= cdf[-1]

    # Interpolate to find radii where CDF reaches 68% and 95%
    # We restrict to strictly increasing region to avoid numerical edge cases
    mask_pos = cdf > 0
    cdf_valid = cdf[mask_pos]
    rad_valid = rad_mid[mask_pos]

    r68 = np.interp(0.68, cdf_valid, rad_valid)
    r95 = np.interp(0.95, cdf_valid, rad_valid)

    # Convert back to degrees
    r68_list.append(r68 * 180.0 / np.pi)
    r95_list.append(r95 * 180.0 / np.pi)

r68_arr = np.array(r68_list)
r95_arr = np.array(r95_list)

fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(energies.to_value("TeV"), r68_arr, label="68% containment")
ax.plot(energies.to_value("TeV"), r95_arr, label="95% containment")

ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("Energy [TeV]")
ax.set_ylabel("Containment radius [deg]")
ax.set_title("PSF containment radii vs energy")
ax.legend()
plt.tight_layout()
plt.show()
```

# Energy dispersion (EDISP)

In questa sezione costruiamo un file FITS **GADF** `EDISP_2D` a partire da un CSV che contiene una matrice di risposta in energia.

**Assunzioni del tuo CSV**
- colonne con `log10(nu_E [GeV]) low/center/high`
- colonne con `log10(reco_E [GeV]) low/center/high`
- ultima colonna che nel tuo esempio è etichettata `dP/dlog10(nu_E [GeV])`, ma (dato che per ogni bin in energia vera scorri i bin in energia ricostruita) viene interpretata come **densità** `dP/dlog10(E_reco)`.

Il codice sotto:
- costruisce gli edges in energia vera (GeV)
- costruisce l’asse **MIGRA = E_reco / E_true**
- converte la densità in probabilità per bin e normalizza per ogni bin di energia vera
- scrive un FITS GADF con un solo bin in **THETA** (0..π), come fatto per AEFF.


```python
import numpy as np
import pandas as pd
from astropy.io import fits

def build_edisp_from_csv(file_csv: str, file_fits: str, theta_single_bin: bool = True):
    """
    Build GADF EDISP_2D from your CSV:
      log10(nu_E) low/center/high,
      log10(reco_E) low/center/high,
      dP/dlog10(E_reco)  (your column is mislabelled in the example)

    Output: FITS with one THETA bin (0..pi) by default.

    Returns
    -------
    matrix : ndarray
        Shape (n_theta, n_migra, n_energy_true), normalized per true-energy bin.
    e_true_edges : ndarray
        True-energy bin edges [GeV]
    migra_edges : ndarray
        Migra bin edges [dimensionless]
    theta_edges : ndarray
        Theta bin edges [rad]
    """

    df = pd.read_csv(file_csv)

    # --- rename to sane internal names ---
    df = df.rename(columns={
        "log10(nu_E [GeV]) low":    "log10Et_lo",
        "log10(nu_E [GeV]) center": "log10Et_c",
        "log10(nu_E [GeV]) high":   "log10Et_hi",
        "log10(reco_E [GeV]) low":    "log10Er_lo",
        "log10(reco_E [GeV]) center": "log10Er_c",
        "log10(reco_E [GeV]) high":   "log10Er_hi",
        # Your CSV label: we interpret as dP/dlog10(E_reco)
        "dP/dlog10(nu_E [GeV])":    "dP_dlog10Er",
    })

    required = ["log10Et_lo","log10Et_c","log10Et_hi","log10Er_lo","log10Er_hi","dP_dlog10Er"]
    missing = [c for c in required if c not in df.columns]
    if missing:
        raise ValueError(f"Missing columns in EDISP CSV: {missing}")

    # --- True energy edges (GeV) ---
    log10Et_edges = np.sort(np.unique(np.concatenate([df["log10Et_lo"].values, df["log10Et_hi"].values])))
    e_true_edges = 10.0 ** log10Et_edges
    nE = len(e_true_edges) - 1

    # --- Build common MIGRA bins from the union over all rows ---
    # For each row: log10(migra) = log10(Ereco) - log10(Etrue_center)
    log10m_lo = df["log10Er_lo"].values - df["log10Et_c"].values
    log10m_hi = df["log10Er_hi"].values - df["log10Et_c"].values

    log10m_edges = np.sort(np.unique(np.concatenate([log10m_lo, log10m_hi])))
    migra_edges = 10.0 ** log10m_edges
    nM = len(migra_edges) - 1

    # --- Convert density to probability per reco-log10 bin ---
    dlog10Er = df["log10Er_hi"].values - df["log10Er_lo"].values
    p_bin = df["dP_dlog10Er"].values * dlog10Er  # unitless probability mass in that reco bin

    # Build matrix as probability per migra bin, normalized per true energy
    # Shape: (theta, migra, energy_true)
    matrix = np.zeros((1, nM, nE), dtype=float)

    # Map energy bins
    e_pairs = (
        df[["log10Et_lo","log10Et_hi"]]
        .drop_duplicates()
        .sort_values(["log10Et_lo","log10Et_hi"])
        .to_numpy()
    )
    ekey = {(float(lo), float(hi)): i for i, (lo, hi) in enumerate(e_pairs)}

    # Map migra bins (rounded keys)
    def _rk(x):
        return float(np.round(x, 12))

    m_pairs = np.vstack([log10m_lo, log10m_hi]).T
    m_pairs_unique = np.unique(np.round(m_pairs, 12), axis=0)
    m_pairs_unique = m_pairs_unique[np.argsort(m_pairs_unique[:, 0])]
    mkey = {(_rk(lo), _rk(hi)): i for i, (lo, hi) in enumerate(m_pairs_unique)}

    for (row, p, lmlo, lmhi) in zip(df.itertuples(index=False), p_bin, log10m_lo, log10m_hi):
        e_idx = ekey[(float(row.log10Et_lo), float(row.log10Et_hi))]
        m_idx = mkey[(_rk(lmlo), _rk(lmhi))]
        matrix[0, m_idx, e_idx] += float(p)

    # Normalize per true-energy bin (sum over migra = 1)
    s = matrix[0].sum(axis=0)  # (nE,)
    s_safe = np.where(s > 0, s, 1.0)
    matrix[0] /= s_safe[None, :]

    # --- THETA axis (single bin) ---
    if theta_single_bin:
        theta_edges = np.array([0.0, np.pi], dtype=float)
        nT = 1
    else:
        raise NotImplementedError("Set theta_single_bin=True (single theta bin 0..pi).")

    # --- Write FITS (GADF) ---
    hdu0 = fits.PrimaryHDU()

    col1 = fits.Column(name="ENERG_LO", format=f"{nE}E", unit="GeV", array=[e_true_edges[:-1]])
    col2 = fits.Column(name="ENERG_HI", format=f"{nE}E", unit="GeV", array=[e_true_edges[1:]])
    col3 = fits.Column(name="MIGRA_LO", format=f"{nM}E", array=[migra_edges[:-1]])
    col4 = fits.Column(name="MIGRA_HI", format=f"{nM}E", array=[migra_edges[1:]])
    col5 = fits.Column(name="THETA_LO", format=f"{nT}E", unit="rad", array=[theta_edges[:-1]])
    col6 = fits.Column(name="THETA_HI", format=f"{nT}E", unit="rad", array=[theta_edges[1:]])

    # Store matrix with dim order (ENERG, MIGRA, THETA) inside the column
    mat_for_fits = matrix.transpose(2, 1, 0)  # (nE, nM, nT)

    col7 = fits.Column(
        name="MATRIX",
        format=f"{nE*nM*nT}D",
        dim=f"({nE},{nM},{nT})",
        array=[mat_for_fits],
    )

    hdu1 = fits.BinTableHDU.from_columns([col1, col2, col3, col4, col5, col6, col7])
    hdu1.header["EXTNAME"]  = "ENERGY DISPERSION"
    hdu1.header["HDUDOC"]   = "https://github.com/open-gamma-ray-astro/gamma-astro-data-formats"
    hdu1.header["HDUVERS"]  = "0.2"
    hdu1.header["HDUCLASS"] = "GADF"
    hdu1.header["HDUCLAS1"] = "RESPONSE"
    hdu1.header["HDUCLAS2"] = "EDISP"
    hdu1.header["HDUCLAS3"] = "FULL-ENCLOSURE"
    hdu1.header["HDUCLAS4"] = "EDISP_2D"

    fits.HDUList([hdu0, hdu1]).writeto(file_fits, overwrite=True)
    print(f"Wrote EDISP FITS: {file_fits}")

    return matrix, e_true_edges, migra_edges, theta_edges
```

# Point Spread Function (PSF)

Qui costruiamo un file FITS **GADF** `PSF_2D_TABLE` a partire da un CSV che contiene:
- `log10(nu_E [GeV]) low/center/high`
- `log10(psi [degrees])` (punti campionati in log10 dell’angolo)
- `dP/dOmega` (densità per steradiante, tipicamente `sr^-1`)

Il codice:
- ricostruisce i bin in **RAD** (in rad) a partire dai punti campionati
- riempie `RPSF` costante per bin
- normalizza (per sicurezza) in modo che ∫ RPSF dΩ = 1 per ogni energia
- scrive un FITS GADF con un solo bin in **THETA** (0..π).


```python
import numpy as np
import pandas as pd
from astropy.io import fits

def build_psf_from_csv(file_csv: str, file_fits: str, theta_single_bin: bool = True):
    """
    Build GADF PSF_2D_TABLE from your CSV:
      log10(nu_E) low/center/high,
      log10(psi [degrees]) (sample points),
      dP/dOmega (sr^-1)

    Output: FITS with one THETA bin (0..pi) by default.

    Returns
    -------
    rpsf : ndarray
        Shape (n_theta, n_rad, n_energy_true), normalized per true-energy bin.
    e_true_edges : ndarray
        True-energy edges [GeV]
    rad_edges : ndarray
        Radial-offset edges [rad]
    theta_edges : ndarray
        Theta edges [rad]
    """

    df = pd.read_csv(file_csv)

    df = df.rename(columns={
        "log10(nu_E [GeV]) low":    "log10Et_lo",
        "log10(nu_E [GeV]) center": "log10Et_c",
        "log10(nu_E [GeV]) high":   "log10Et_hi",
        "log10(psi [degrees])":     "log10psi_deg",
        "dP/dOmega":                "rpsf",
    })

    required = ["log10Et_lo","log10Et_hi","log10psi_deg","rpsf"]
    missing = [c for c in required if c not in df.columns]
    if missing:
        raise ValueError(f"Missing columns in PSF CSV: {missing}")

    # --- Energy edges (GeV) ---
    log10Et_edges = np.sort(np.unique(np.concatenate([df["log10Et_lo"].values, df["log10Et_hi"].values])))
    e_true_edges = 10.0 ** log10Et_edges
    nE = len(e_true_edges) - 1

    # --- RAD bins from sampled log10(psi[deg]) points ---
    log10psi = np.sort(np.unique(df["log10psi_deg"].values))
    if len(log10psi) < 2:
        raise ValueError("Need at least 2 psi sample points to build RAD bin edges.")

    d = np.diff(log10psi)
    step = float(np.median(d))
    # edges in log10 space (assume ~uniform grid)
    log10psi_edges = np.concatenate([[log10psi[0] - 0.5*step], 0.5*(log10psi[:-1]+log10psi[1:]), [log10psi[-1] + 0.5*step]])
    psi_deg_edges = 10.0 ** log10psi_edges
    rad_edges = np.deg2rad(psi_deg_edges)
    nR = len(rad_edges) - 1

    # --- THETA axis ---
    if theta_single_bin:
        theta_edges = np.array([0.0, np.pi], dtype=float)
        nT = 1
    else:
        raise NotImplementedError("Set theta_single_bin=True (single theta bin 0..pi).")

    # Allocate (theta, rad, energy)
    rpsf = np.zeros((1, nR, nE), dtype=float)

    # Map energy bins
    e_pairs = (
        df[["log10Et_lo","log10Et_hi"]]
        .drop_duplicates()
        .sort_values(["log10Et_lo","log10Et_hi"])
        .to_numpy()
    )
    ekey = {(float(lo), float(hi)): i for i, (lo, hi) in enumerate(e_pairs)}

    # Map rad bins: each row provides a sample at log10psi_deg -> choose nearest bin center
    rad_centers = 0.5*(rad_edges[:-1] + rad_edges[1:])
    log10rad_centers_deg = np.log10(np.rad2deg(rad_centers))

    # Fill: for each energy bin, assign rpsf for each sampled psi to nearest radial bin
    for row in df.itertuples(index=False):
        e_idx = ekey[(float(row.log10Et_lo), float(row.log10Et_hi))]
        j = int(np.argmin(np.abs(log10rad_centers_deg - float(row.log10psi_deg))))
        rpsf[0, j, e_idx] = float(row.rpsf)

    # --- Normalize per energy bin so that integral over solid angle is 1 ---
    # dOmega for annulus: 2pi (cos(r_lo) - cos(r_hi))
    r_lo = rad_edges[:-1]
    r_hi = rad_edges[1:]
    dOmega = 2.0*np.pi*(np.cos(r_lo) - np.cos(r_hi))  # (nR,)

    integ = (rpsf[0] * dOmega[:, None]).sum(axis=0)  # (nE,)
    integ_safe = np.where(integ > 0, integ, 1.0)
    rpsf[0] /= integ_safe[None, :]

    # --- Write FITS (GADF) ---
    hdu0 = fits.PrimaryHDU()

    col1 = fits.Column(name="ENERG_LO", format=f"{nE}E", unit="GeV", array=[e_true_edges[:-1]])
    col2 = fits.Column(name="ENERG_HI", format=f"{nE}E", unit="GeV", array=[e_true_edges[1:]])
    col3 = fits.Column(name="THETA_LO", format=f"{nT}E", unit="rad", array=[theta_edges[:-1]])
    col4 = fits.Column(name="THETA_HI", format=f"{nT}E", unit="rad", array=[theta_edges[1:]])
    col5 = fits.Column(name="RAD_LO", format=f"{nR}E", unit="rad", array=[rad_edges[:-1]])
    col6 = fits.Column(name="RAD_HI", format=f"{nR}E", unit="rad", array=[rad_edges[1:]])

    # Store as (ENERG, RAD, THETA)
    rpsf_for_fits = rpsf.transpose(2, 1, 0)  # (nE, nR, nT)

    col7 = fits.Column(
        name="RPSF",
        format=f"{nE*nR*nT}D",
        dim=f"({nE},{nR},{nT})",
        unit="sr-1",
        array=[rpsf_for_fits],
    )

    hdu1 = fits.BinTableHDU.from_columns([col1,col2,col3,col4,col5,col6,col7])
    hdu1.header["EXTNAME"]  = "POINT SPREAD FUNCTION"
    hdu1.header["HDUDOC"]   = "https://github.com/open-gamma-ray-astro/gamma-astro-data-formats"
    hdu1.header["HDUVERS"]  = "0.2"
    hdu1.header["HDUCLASS"] = "GADF"
    hdu1.header["HDUCLAS1"] = "RESPONSE"
    hdu1.header["HDUCLAS2"] = "PSF"
    hdu1.header["HDUCLAS3"] = "FULL-ENCLOSURE"
    hdu1.header["HDUCLAS4"] = "PSF_2D_TABLE"

    fits.HDUList([hdu0, hdu1]).writeto(file_fits, overwrite=True)
    print(f"Wrote PSF FITS: {file_fits}")

    return rpsf, e_true_edges, rad_edges, theta_edges
```
