### Reading the IRFs from files

```python
import numpy as np
from astropy.io import fits

def read_aeff_with_edges(fits_file):
    """Read AEFF_2D FITS and return bin edges, centers, and values."""
    with fits.open(fits_file) as hdul:
        data = hdul[1].data
        e_lo = data["ENERG_LO"][0]      # length n_energy
        e_hi = data["ENERG_HI"][0]
        theta_lo = data["THETA_LO"][0]  # length n_theta
        theta_hi = data["THETA_HI"][0]
        aeff = data["EFFAREA"][0]       # shape (n_energy, n_theta)

    # Bin centers
    e_center = np.sqrt(e_lo * e_hi)                # geometric mean for energy
    theta_center = 0.5 * (theta_lo + theta_hi)    # arithmetic mean for offset/zenith

    # Bin edges
    e_edges = np.hstack([e_lo, e_hi[-1]])
    theta_edges = np.hstack([theta_lo, theta_hi[-1]])

    return e_edges, e_center, theta_edges, theta_center, aeff

# --- Read AEFF IRFs ---
e_edges, e_center, theta_edges, theta_center, aeff_value = read_aeff_with_edges("../aeff.fits")
e_edges_nocuts, e_center_nocuts, theta_edges_nocuts, theta_center_nocuts, aeff_nocuts_value = read_aeff_with_edges("../aeff_nocuts.fits")

# --- Use bin centers as you did before ---
km3_e_binc = e_center

# --- Optional: print content ---
print("Effective area (shape):", aeff_value.shape)
print("Energy bin centers:", km3_e_binc)
```

## The PSF

```python
from gammapy.irf import PSF3D
import numpy as np

# Read PSF
km3_psf = PSF3D.read("psf.fits")

# Load LoI reference
km3_psf_loi = np.loadtxt('LoI/PSF_LoI.csv', delimiter=';', skiprows=0)

# Compute energy bin centers (geometric mean) in GeV
energy_axis = km3_psf.axes["energy_true"]   # QuantityAxis
km3_e1_binc = np.sqrt(energy_axis.edges[:-1] * energy_axis.edges[1:])  # edges in GeV
km3_e1_binc = km3_e1_binc.to('GeV').value  # convert to GeV if needed
```

```python
import numpy as np
import astropy.units as u
from scipy.interpolate import interp1d
from gammapy.irf import PSF3D

# --- Read PSF ---
km3_psf = PSF3D.read("psf.fits")

psf_values = np.array(km3_psf.data.data)  # shape (energy, offset, rad)
rad_axis = km3_psf.axes["rad"]
offset_axis = km3_psf.axes["offset"]

rad_center = 0.5 * (rad_axis.edges[:-1] + rad_axis.edges[1:])  # Quantity

# --- Choose offsets > 90 deg, if applicable ---
mask_offset = (offset_axis.center > 90 * u.deg)  # Convert Quantity to bool
psf_selected = psf_values[:, mask_offset, :]  # shape (energy, selected_offsets, rad)

# --- Average over selected offsets ---
psf_avg = np.mean(psf_selected, axis=1)  # shape (energy, rad)

# --- Radial bin areas in sr ---
rad_edges_rad = rad_axis.edges.to("rad").value
sizes_rad_bins_sr = np.pi * (rad_edges_rad[1:]**2 - rad_edges_rad[:-1]**2)

rad_drad = np.cumsum(psf_avg * sizes_rad_bins_sr[None, :], axis=1)
# check integral before normalization
print("Before normalization:", rad_drad[:, -1])
# normalize so each energy integrates to 1
rad_drad /= rad_drad[:, -1][:, None]
print("After normalization:", rad_drad[:, -1])


# --- Interpolate containment ---
km3_psf_cr50, km3_psf_cr68, km3_psf_cr95 = [], [], []

for e_idx in range(rad_drad.shape[0]):
    interp = interp1d(rad_drad[e_idx, :], rad_center.to(u.deg).value, assume_sorted=True)
    km3_psf_cr50.append(interp(0.50))
    km3_psf_cr68.append(interp(0.68))
    km3_psf_cr95.append(interp(0.95))

km3_psf_cr50 = np.array(km3_psf_cr50)
km3_psf_cr68 = np.array(km3_psf_cr68)
km3_psf_cr95 = np.array(km3_psf_cr95)

print("CR50:", km3_psf_cr50)


print(km3_psf_cr50)






```

```python
### We interpolate the rad values at the corresponding fractions of the integral
km3_psf_cr50=[]
km3_psf_cr68=[]
km3_psf_cr95=[]
rad_center = 0.5 * (rad_axis.edges[:-1] + rad_axis.edges[1:])  # rad_axis.edges is already a Quantity
for e in range(rad_drad.shape[0]):    # cumulative PSF along rad axis for this energy
    cumulative = rad_drad[e, :]  # length n_rad
    
    # interpolate: cumulative vs radius
    interp_containment = interp1d(cumulative, rad_center.to(u.deg).value, assume_sorted=True)
    km3_psf_cr50.append(interp_containment(0.50))
    km3_psf_cr68.append(interp_containment(0.68))
    km3_psf_cr95.append(interp_containment(0.95))
```

## The energy dispersion

```python
import numpy as np
import astropy.units as u
from gammapy.irf import EnergyDispersion2D

# --- Lettura dell’IRF ---
edisp_km3net = EnergyDispersion2D.read("edisp.fits", hdu="EDISP_2D")

# --- Assi ---
energy_axis = edisp_km3net.axes["energy_true"]   # true energy axis
migra_axis  = edisp_km3net.axes["migra"]         # migration axis
offset_axis = edisp_km3net.axes["offset"]        # offset axis (gradi o radianti)

# --- Centri dei bin di migrazione ---
migra_center_km3 = np.sqrt(migra_axis.edges[1:] * migra_axis.edges[:-1])
print("migra_axis.edges:", migra_axis.edges)

# --- Dati interni come array numpy ---
# In Gammapy 2.0 l’attributo è `edisp_km3net.data.data` → MapData (ndarray)
# --- Dati interni come array numpy ---
data = np.array(edisp_km3net.data.data)  # converte da memoryview → ndarray

# --- Media su offset (ad es. zenith > 90°) ---
mean_over_offset = np.mean(data[..., 3:], axis=2)

# --- Cumulativa lungo l'asse migra ---
mu_dmu_km3 = np.cumsum(mean_over_offset, axis=1)

print("mu_dmu_km3[:,-1]:", mu_dmu_km3[:, -1])




```

```python
### We interpolate the migra values at the corresponding fractions of the integral
km3_edisp_cr50=[]
km3_edisp_cr10=[]
km3_edisp_cr90=[]
for i in range(mu_dmu_km3.shape[0]):
    interp_containment = interp1d(mu_dmu_km3[i,:], migra_center_km3, assume_sorted=True)
    km3_edisp_cr50.append(interp_containment(0.50))
    km3_edisp_cr10.append(interp_containment(0.10))
    km3_edisp_cr90.append(interp_containment(0.90))
```

```python
edisp_cta = cta_irfs['edisp']
```

```python
migra_center_cta=edisp_cta.data.axes[1].center
energy_center_cta=edisp_cta.data.axes[0].center
energy_mask_cta = (energy_center_cta<250*u.TeV) & (energy_center_cta>0.1*u.TeV)
mu_dmu_cta = np.cumsum(edisp_cta.data.data[energy_mask_cta,:,1],axis=1)
```

```python
### We interpolate the migra values at the corresponding fractions of the integral
cta_edisp_cr50=[]
cta_edisp_cr10=[]
cta_edisp_cr90=[]
for i in range(energy_mask_cta.sum()):
    interp_containment = interp1d(mu_dmu_cta[i,:], migra_center_cta, assume_sorted=True)
    cta_edisp_cr50.append(interp_containment(0.50))
    cta_edisp_cr10.append(interp_containment(0.10))
    cta_edisp_cr90.append(interp_containment(0.90))
```

```python
# Paper plot
plot_utils.mpl_settings()
from matplotlib.lines import Line2D

fig_width = 8.8 / 2.54
ax_width = 0.83
fig_height = 2.2
ax_height = fig_width * ax_width / 1.618 / fig_height  # golden ratio

fig = plt.figure(figsize=(fig_width, fig_height), dpi=300)
ax = fig.add_axes([0.16, 0.18, ax_width, ax_height])
ax.set_xlabel('$E\,[\mathrm{TeV}]$')
ax.set_ylabel('$E_\mathrm{reco}/E_\mathrm{true}$')
ax.set_xscale('log')
ax.set_yscale('log')

ax.plot(km3_e1_binc/1e3, km3_edisp_cr50, color='tab:blue', label='KM3NeT')
ax.plot(km3_e1_binc/1e3, km3_edisp_cr10, color='tab:blue', ls='--')
ax.plot(km3_e1_binc/1e3, km3_edisp_cr90, color='tab:blue', ls=':')

ax.set_xlim(0.05, 1.5e5)
ax.set_xticks([0.1, 10, 1e3, 1e5])
ax.set_ylim(0.01, 10)

skip = 3
ax.plot(energy_center_cta[energy_mask_cta][::skip], cta_edisp_cr50[::skip], color='tab:red', label='CTA')
ax.plot(energy_center_cta[energy_mask_cta][::skip], cta_edisp_cr10[::skip], color='tab:red', ls='--')
ax.plot(energy_center_cta[energy_mask_cta][::skip], cta_edisp_cr90[::skip], color='tab:red', ls=':')

leg1 = ax.legend(loc='upper right')
ax.legend([Line2D([0], [0], color='0.5', ls='-'), Line2D([0], [0], color='0.5', ls='--'), Line2D([0], [0], color='0.5')],
          ['median', '$90\%$ quantile', '$10\%$ quantile'], loc='lower right')
ax.add_artist(leg1)



plot_utils.format_log_axis(ax.xaxis)
plot_utils.format_log_axis(ax.yaxis)

for form in ['png', 'pdf']:
     fig.savefig('plots/edisp_cta_km3net.{}'.format(form), dpi=300, bbox_inches='tight')
```

## The atmospheric neutrino fluxes 

```python
bkg_nu = Background2D.read("bkg_nu.fits")
bkg_mu = Background2D.read("bkg_mu.fits")
mu_horizon_raw = np.loadtxt('raw_muon_rate.dat')[:,1]
atm_prompt_rate = np.loadtxt('prompt_nu_rate.dat')[:,1]
```

```python
from gammapy.irf import Background2D
import numpy as np
import astropy.units as u

# --- Lettura del file di background ---
# --- Accesso agli assi (nuova sintassi) ---
energy_axis = bkg_nu.axes["energy"]   # o "energy_true" se così definito
theta_axis  = bkg_nu.axes["offset"]   # o "fov_offset" / "theta", dipende dal file

# --- Centri e bordi ---
t_binc_fine      = theta_axis.center.to_value("rad")
e_binc_fine_ext  = energy_axis.center.to_value("GeV")
e_bin_fine_ext   = energy_axis.edges.to_value("MeV")

# --- Calcoli ---
de_domega_reco = 4 * np.pi / len(t_binc_fine) * np.diff(e_bin_fine_ext)
t_mask = t_binc_fine * 180 / np.pi > 80  # seleziona regione di analisi
t_year = 365 * 24 * 3600  # secondi in un anno

print("t_binc_fine:", t_binc_fine)
print("e_binc_fine_ext:", e_binc_fine_ext)
```

```python
atm_conv_rate = np.array(bkg_nu.data.data)  # shape: (n_energy, n_theta)
mu_horizon_smoothed = np.array(bkg_mu.data.data)[:, 5]  # shape: (n_energy,)
```

```python
plot_utils.mpl_settings()
fig_width = 8.8 / 2.54
ax_width = 0.86
fig_height = 2.6
ax_height = fig_width * ax_width / 1.5 / fig_height
fig = plt.figure(figsize=(fig_width, fig_height), dpi=300)
ax = fig.add_axes([0.13, 0.15, ax_width, ax_height])

ax.plot(e_binc_fine_ext*1e-3, atm_conv_rate[:,t_mask].sum(axis=1) * de_domega_reco * t_year,
        label='Conv. + Prompt neutrinos', c='red')
ax.plot(e_binc_fine_ext*1e-3, atm_prompt_rate * de_domega_reco * t_year,
        label='Prompt neutrinos', c='green', ls='--')
ax.plot(e_binc_fine_ext*1e-3, mu_horizon_raw * de_domega_reco * t_year,
        '-', label='Muons', c='tab:blue', ds='steps-mid', zorder=0)
ax.plot(e_binc_fine_ext*1e-3, mu_horizon_smoothed * de_domega_reco * t_year,zorder=0, c='k',ls='-', label='Muons, smoothed')

ax.loglog()
ax.set_xlabel('$E_\mathrm{reco}$ [TeV]')
ax.set_ylabel('Background Rate [yr$^{-1}$]')
ax.legend(loc='upper right', fontsize='small')
ax.set_ylim(3e-5, 3e5)
ax.set_xticks([10**i for i in range(-3,6,2)])
plot_utils.format_log_axis(ax.xaxis)
```
