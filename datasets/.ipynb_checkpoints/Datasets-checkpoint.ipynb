## 10) Plot reading: energy distribution

### Quick questions
- In which energy range does the source dominate over background?

### Mini-variation
Overplot the cumulative counts (integral above E) for signal and background.


```python
make_energy_counts_plot(
    energy_axis,
    npred_bkg_nu_all,
    npred_bkg_mu_all,
    npred_src_all,
    npred_sum_all,
    counts_all,
    src_reg_mask,
    SOURCE_NAME,
)
```

```python
def cumulative_above(map_):
    data = map_.data  
    spec = data.sum(axis=tuple(range(1, data.ndim)))  # -> 1D (n_energy,)
    return np.cumsum(spec[::-1])[::-1]

cum_bkg = cumulative_above(npred_bkg_sum_all)
cum_src = cumulative_above(npred_src_all)
e = energy_axis.center.to_value("TeV")

plt.figure()
plt.loglog(e, cum_bkg, label="bkg")
plt.loglog(e, cum_src, label="src")
plt.xlabel("Energy [TeV]")
plt.ylabel("Cumulative counts (>E)")
plt.title("Cumulative counts above E")
plt.legend()
plt.show()
```

## 11) Map reading: morphology

### Quick questions
- What does smoothing do visually and statistically?
- How would you choose a smoothing scale compared to the PSF?

### Mini-variations
A. Change smoothing kernel (0.1°, 0.25°, 0.5°) and compare.  
B. Plot a 1D profile through the source (slice in RA or Dec).


```python
print("===> Creating smoothed count map...")
fig = plt.figure(figsize=(3.5, 2.75), dpi=300)
ax = counts_all.sum_over_axes().smooth(0.25 * u.deg).plot(
    add_cbar=False,
    vmin=0,
    cmap="viridis",
)
cbar = ax.figure.colorbar(ax.images[0], ax=ax)
cbar.set_label("Counts per pixel")

src_reg_pix = src_region.to_pixel(ax.wcs)
src_reg_pix.plot(ax=ax, ec="w", ls="--", lw=1)

ax.set_title("Smoothed Count Map")
plt.subplots_adjust(left=0.15, right=0.95, bottom=0.17, top=0.97)

ax.figure.savefig(f"plots/countmap_{SOURCE_NAME}.pdf",
                  format="pdf",
                  bbox_inches="tight")
ax.figure.savefig(f"plots/countmap_{SOURCE_NAME}.png",
                  format="png",
                  dpi=300,
                  bbox_inches="tight")
plt.show()
```
